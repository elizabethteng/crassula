## make models

```python
# setting star parameters (from make_Grid.py)

filename="../old/hdf5/29e.hdf5"
pdfname="../visualization/seds_29e.pdf"

T_star = 4000
logL_star = 0.

logM_disk = -6.
logR_disk = 1.
h_0 = 0.1

logR_in = 0
gamma = 1.
beta = 1.1

logM_env = -6.
logR_env = 3
f_cav = 0.5
ksi = 1.1

loga_max = 1.
p = 3. 

incl = 45.

```

```python
# from function run_yso_model in make_Grid.py

# Set up the dust properties.
dust_gen = dust.DustGenerator(dust.__path__[0]+"/data/diana_wice.hdf5")
ddust = dust_gen(10.**loga_max / 1e4, p)
env_dust_gen = dust.DustGenerator(dust.__path__[0]+\
        "/data/diana_wice.hdf5")
edust = env_dust_gen(1.0e-4, 3.5)

# Calculate alpha correctly.
alpha = gamma + beta

# Fix the scale height of the disk.
h_0 *= (10.**logR_disk)**beta
```

```python
# Set up the model.
model = modeling.YSOModel()
```

```python
model.add_star(luminosity=10.**logL_star, temperature=T_star)
```

```python
model.set_spherical_grid(10.**logR_in, 10.**logR_env, 100, 101, 2, \
        code="radmc3d")
```

```python
model.add_pringle_disk(mass=10.**logM_disk, rmin=10.**logR_in, \
        rmax=10.**logR_disk, plrho=alpha, h0=h_0, plh=beta, dust=ddust)
```

```python
model.add_ulrich_envelope(mass=10.**logM_env, rmin=10.**logR_in, \
        rmax=10.**logR_env, cavpl=ksi, cavrfact=f_cav, dust=edust)
```

```python
model.grid.set_wavelength_grid(0.1,1.0e5,500,log=True)
```

```python
# Run the thermal simulation
model.run_thermal(code="radmc3d", nphot=1e6, \
        modified_random_walk=True, verbose=False, setthreads=4, \
        timelimit=10800)
```

```python
# Run the SED
model.set_camera_wavelength(numpy.logspace(-1.,4.,500))
```

```python
model.run_sed(name="SED", nphot=1e5, loadlambda=True, incl=incl, \
        pa=0., dpc=140., code="radmc3d", camera_scatsrc_allfreq=True, \
        verbose=False, setthreads=4)
```

```python
# run the image
model.run_image(name="1mm", nphot=1e5, npix=512, \
        pixelsize=0.05, lam="1000", incl=0., pa=0.,\
        dpc=140, verbose=True)

# pixel is in units of arcsec, to match observations
# lambda is in units of microns, so the above makes an image at 1mm. 
```

```python
# Write out the file.
model.write_yso(filename)
```

## plot models

```python
# From plot_models.py

# Open up a pdf file to plot into.
pdf = PdfPages(pdfname)

# read in the data
model = modeling.YSOModel()
model.read_yso(filename)
```

```python
# get the temperature structure
r = model.grid.r
theta = model.grid.theta

rr, tt = numpy.meshgrid(r,theta,indexing = "ij")
R = rr * numpy.sin(tt)
z = rr * numpy.cos(tt)

temperature = model.grid.temperature[1][:, :, 0]
```

```python
#Plots
fig, ax = plt.subplots(nrows=1, ncols=2)

# Plot the temperature
#ax[0].contourf(R, z, temperature, 100)

# plot the image
ax[0].imshow(model.images["1mm"].image[:,:,0,0])

# Plot the SED.
line = ax[1].loglog(model.spectra["SED"].wave, model.spectra["SED"].flux, \
        "-")
```

```python
# Adjust the figure.

ax[1].set_ylim(1.0e-6,1.0e5)
ax[0].set_xlim(0,100)
ax[0].set_ylim(0,100)
fig.set_size_inches((12.8,4.8))
fig.suptitle(filename.replace("_"," ").replace("beta","\n beta"))

# Now save the figure.
pdf.savefig(fig)
plt.clf()

# close pdf
pdf.close()
```

```python
# stacked plot
pdfname="../visualization/SED/stack5.pdf"

files=[["../old/hdf5/29a.hdf5","T=4000, logR_disk=1., loga_max=1, logL_star=0"],\
       ["../old/hdf5/29b.hdf5","logR_disk=2."],\
       ["../old/hdf5/29c.hdf5","T=3000"],\
       ["../old/hdf5/29d.hdf5","loga_max=4"],\
       ["../old/hdf5/29e.hdf5","logL_star=3"]]

#steps=numpy.linspace(0.0,0.5,len(files))
steps=numpy.linspace(0.0,0.7,7)
colors= plt.cm.Set1(steps)

pdf = PdfPages(pdfname)

fig,ax=plt.subplots(1,figsize=(12,9))

for i in range(len(files)):
    model = modeling.YSOModel()
    model.read_yso(files[i][0])
    ax.loglog(model.spectra["SED"].wave, model.spectra["SED"].flux, "-",color=colors[i], label=files[i][1])

ax.set_ylim(1.0e-6,1.0e5)

plt.legend()
plt.show()

# Now save the figure.
pdf.savefig(fig)
plt.clf()

# close pdf
pdf.close()
```
